```python
original_sheet_path = "/Users/kevinxie/Desktop/LLM CoT/LLM-CoT/Clinical Benchmark and LLM.xlsx"

def load_sheet(path, sheet_name):
    sheet = pd.ExcelFile(path)

    data = sheet.parse(sheet_name)

    return data

sheet = load_sheet(original_sheet_path, "Task-all")
```

```python
def get_task_to_type_mapping(sheet):

    type_map= {}

    mp = {
        "Event Extraction": "ext",
        "Named Entity Recognition": "ext",
        "Natural Language Inference": "clf",
        "Normalization and Coding": "clf",
        "Question Answering": "gen",
        "Semantic Similarity": "clf",
        "Summarization": "gen",
        "Text Classification": "clf"
    }
    category_map = {}

    for row_idx, task_name in enumerate(sheet['Task-Original']):
        type_map[task_name] = sheet["Task Type - fine grained"][row_idx]


    for task, type in type_map.items():
        category_map[task] = mp[type]
    

    return type_map, category_map

type_map, category_map = get_task_to_type_mapping(sheet)
```

```python
def get_metrics(json_path, task_name):
    """
    Given the path to a json file, return all necessary metrics:

    [model name]
    [task name]
    [task type]
    [direct score] or [cot score]

    "rouge, "f1_event", "accuracy"
    """

    model_names = []
    task_names = []
    task_types = []
    scores = []

    with open(json_path, 'r') as f:
        data = json.load(f)

    for d_model, d_metrics in data.items():
        model_names.append(d_model)

        task_names.append(task_name)
        task_types.append(type_map[task_name])

        if "rouge" in d_metrics:
            scores.append(d_metrics["rouge"]["mean"])

        elif "f1_event" in d_metrics:
            scores.append(d_metrics["f1_event"]["mean"])

        elif "accuracy" in d_metrics:
            scores.append(d_metrics["accuracy"]["mean"])

        else:
            print('nope')

    return model_names, task_names, task_types, scores

def get_difference(cot_score, direct_score):
    return round(direct_score - cot_score, 2)

def get_relative_difference(cot_score, direct_score):
    if direct_score == 0:
        return 0
    return round(((direct_score - cot_score) / direct_score) * 100, 2)


def create_sheet(all_tasks, prompt_num):
    output = {
        "Model Name": [],
        "Task Name": [],
        "Task Type": [],
        "Direct Score": [],
        "CoT Score": [],
        "Difference": [],
        "Relative Difference": []
    }

    for task_file_name in all_tasks:
        prompt_n = task_file_name.split("prompt_")
        prompt_n = int(prompt_n[1][0])   # [0, 3]

        task_name = task_file_name.split(".prompt")[0]

        if prompt_n == prompt_num:  # consider this prompt
            prompt_type = task_file_name.split("prompt")
            prompt_type = prompt_type[1].split(".")[1]   # cot or direct

            json_path = os.path.join("/Users/kevinxie/Desktop/LLM CoT/LLM-CoT/performance_4", task_file_name)

            model_names, task_names, task_types, scores = get_metrics(json_path, task_name)

            if prompt_type == "cot":
                output["Model Name"].extend(model_names)
                output["Task Name"].extend(task_names)
                output["Task Type"].extend(task_types)
                output["CoT Score"].extend(scores)

            elif prompt_type == "direct":
                output["Direct Score"].extend(scores)

            else:
                print("INVALID PROMPT TYPE")

    for idx, direct_score in enumerate(output["Direct Score"]):
        # output["Difference"].append(ca)
        diff = get_difference(output["CoT Score"][idx], output["Direct Score"][idx])

        rel_diff = get_relative_difference(output["CoT Score"][idx], output["Direct Score"][idx])

        output["Difference"].append(diff)
        output["Relative Difference"].append(rel_diff)


    return output

def dict_to_csv(d, output_name):
    # Convert to DataFrame
    df = pd.DataFrame(d)

    # Save to CSV
    df.to_csv(output_name, index=False)

    print("Successfully Saved at ", output_name)


prompt_1_sheet = create_sheet(all_tasks, 0)
prompt_2_sheet = create_sheet(all_tasks, 1)
prompt_3_sheet = create_sheet(all_tasks, 2)
prompt_4_sheet = create_sheet(all_tasks, 3)


dict_to_csv(prompt_1_sheet, "prompt_1.csv")
dict_to_csv(prompt_2_sheet, "prompt_2.csv")
dict_to_csv(prompt_3_sheet, "prompt_3.csv")
dict_to_csv(prompt_4_sheet, "prompt_4.csv")
```

```python
def get_task_name_performance(google_sheet, sheet_name, is_CoT):
    task_performances = {}  # maps each TASK NAME (i.e. ADE-Identification) to their performances

   
    data = google_sheet.parse(sheet_name)

    for row_idx, task_name in enumerate(data["Task Name"]):
        if task_name not in task_performances:
            task_performances[task_name] = []

        if is_CoT:
            score = data["CoT Score"][row_idx]

        else:
            score = data["Direct Score"][row_idx]

        task_performances[task_name].append(score)

    # Average the performances
    for task, values_list in task_performances.items():
        task_performances[task] = round(sum(values_list) / len(values_list), 2)

    return task_performances


google_sheet = pd.ExcelFile("/Users/kevinxie/Desktop/LLM CoT/LLM-CoT/New-Prompt-Sheet.xlsx")

# Get performances on each task via CoT prompting
cot_task_performances_1 = get_task_name_performance(google_sheet, "prompt_1", is_CoT=True)
direct_task_performances_1 = get_task_name_performance(google_sheet, "prompt_1", is_CoT=False)

cot_task_performances_2 = get_task_name_performance(google_sheet, "prompt_2", is_CoT=True)
direct_task_performances_2 = get_task_name_performance(google_sheet, "prompt_2", is_CoT=False)

cot_task_performances_3 = get_task_name_performance(google_sheet, "prompt_3", is_CoT=True)
direct_task_performances_3 = get_task_name_performance(google_sheet, "prompt_3", is_CoT=False)

cot_task_performances_4 = get_task_name_performance(google_sheet, "prompt_4", is_CoT=True)
direct_task_performances_4 = get_task_name_performance(google_sheet, "prompt_4", is_CoT=False)
```

```python
def create_performances_csv(cot_task_performances, direct_task_performances, prompt_number):

    final_csv = {
        "Task Name": [],
        "Task Type": [],
        "Direct Score": [],
        "CoT Score": [],
        "Difference": [],
        "Relative Difference": []
    } 

    for task_name, score in cot_task_performances.items():
        final_csv["Task Name"].append(simplified_task_name_map[task_name])
        final_csv["Task Type"].append(type_map[task_name])
        final_csv["CoT Score"].append(score)

    for task_name, score in direct_task_performances.items():
        final_csv["Direct Score"].append(score)

    cot_scores = list(cot_task_performances.values())
    direct_scores = list(direct_task_performances.values())

    for direct, cot in zip(direct_scores, cot_scores):
        final_csv["Difference"].append(round(direct - cot, 2))
        final_csv["Relative Difference"].append(round((direct - cot) / direct * 100, 2))

    df = pd.DataFrame(final_csv)
    df.to_csv(f"Task Name vs. Performance.prompt_{prompt_number}.csv")

    print(f"Successfully saved at 'Task Name vs. Performance.prompt_{prompt_number}.csv'")

create_performances_csv(cot_task_performances_1, direct_task_performances_1, 1)
create_performances_csv(cot_task_performances_2, direct_task_performances_2, 2)
create_performances_csv(cot_task_performances_3, direct_task_performances_3, 3)
create_performances_csv(cot_task_performances_4, direct_task_performances_4, 4)
```
